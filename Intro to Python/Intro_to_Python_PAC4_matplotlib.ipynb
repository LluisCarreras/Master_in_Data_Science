### Ejercicio 3
Representad la función que queráis de la forma que consideréis. Sed creativos a la hora de escoger la función que vayáis a representar y el intervalo de valores. **(0.5 puntos)**

Dada la libertad que permite este ejercicio, he considerado que la gráfica que deseo representar es el diagrama de bifurcaciones, una de las gráficas que encontré más sugerentes cuando estudiaba la Licenciatura en Física, más concretamente en la asignatura Sistemas Dinámicos. Esta gráfica sumariza el comportamiento de la ecuación logística, que xiene dada por la ecuación:

x[t+1] = Cx[t](1-x[t])

Es una ecuación iterativa, en la que x[t] es el valor actual de x y x[t+1] es el valor de x en ela siguiente iteración. Esta ecuación es una de las más simples que capturan la esencia del comportamiento caótico. Esta ecuación muestra un comportamiento muy curioso a medida que se va variando el valor de C. 

Construyo una función que simule el comportamiento de la ecuación logística. Por ejemplo, veamos qué ocurre al tomar un valor C=0.5. Necesitamos también empezar con un valor cualquiera para x, que tiene que estar en el rango [0, 1], y en este ejemplo utilizaré el valor x=0.2. Si vamos iterando, obtenemos:

```python
def logistic_equation(C=3, iterations=21, seed=0.2):
    '''
    Función que calcula y representa gráficamente los primeros valores al iterar
    la ecuación logística.
    '''
    # Empezamos con un primer valor para x, que hace las veces de semilla de las
    # posteriores iteraciones.
    x = seed
    xs = [x]
    
    # Se itera varias veces y se guardan los valores obtenidos en una lista.
    for i in range(1, iterations):
        x = C * x * (1 - x)
        xs.append(x)
    
    # Se muestran los valores que ha ido tomando x al ir iterando. 
    print xs
    
    # Se representa gráficamente el valor que va tomando x en las sucesivas iteraciones.
    plt.scatter(range(iterations), xs, color='red', marker='o', s=15)
    plt.plot(range(iterations), xs, color='blue')
    plt.ylim(0, 1)
    plt.xlabel('C')
    plt.ylabel('x')
    title = 'Logistic equation with C=%.3f and x0=%.3f' % (C, seed)
    plt.title(title)
    plt.grid(True)
    plt.show()
 
# Se llama a la función para que muestre la gráfica con C=0.5.    
logistic_equation(0.5, 21, 0.2)
```

Vemos que con C=0.5 x tiende rápidamente a 0. Probemos ahora con un valor más alto para C, por ejemplo, C=1.5:

```python
# Se llama a la función para que muestre la gráfica con C=1.5.  
logistic_equation(1.5, 21, 0.2)
```

Vemos que ahora, con C=1.5, x tiende a 0.33. Este valor al que tiende se llama *punto fijo*. Veamos ahora qué pasa con C=3.2:

```python
# Se llama a la función para que muestre la gráfica con C=3.2.  
logistic_equation(3.2, 21, 0.2)
```

En este caso vemos que no tiende a un punto fijo  sino que se entra en una oscilación entre dos valores, que en este caso aproximadamente son 0.513 y 0.799. Además, resulta que estos valores se alcanzarán cualquiera que sea el valor x0, elvalor inicial de x. Veámoslo, tomando el mismo caso pero con x0=0.85:

```python
# Se llama a la función para que muestre la gráfica con C=3.2 y un valor distinto de x0.  
logistic_equation(3.2, 21, 0.85)
```

Vemos que, efectivamente, converge a la misma oscilación. Este tipo de comportamiento se llama *atractor* ya que, cualquier condición inicial "es atraido" hacia esta oscilación.

Este comportamiento en que converge a una oscilación entre dos valores se da para valores de C entre 3.0 y alrededor de 3.4, aunque los dos valores sobre los que oscila serán diferentes para cada valor de C. Veámoslo:

```python
# Se llama a la función para que muestre la gráfica con C=3.35.  
logistic_equation(3.35, 21, 0.85)
```

Vemos que para C=3.35 el atractor está formado por los dos valores 0.465 y 0.833.

En un valor entre C=3.4 y 3.5 ocurre un cambio abrupto de comportamiento. Dado un valor cualquiera para x0, el sistema finalmente alcanza una oscilación entre no dos sino cuatro valores. Veámoslo:

```python
# Se llama a la función para que muestre la gráfica con C=3.49.  
logistic_equation(3.49, 21, 0.85)
```

Vemos que con C=3.49 el sistema oscila entre los cuatro valores aproximados 0.480, 0.871, 0.391 y 0.832.

Resulta que este comportamiento se va repitiendo a medida que aumenta C, y para ciertos valores de C el número de valores sobre los que oscila se dobla, llegando a una situación en la que para C=3.569946 oscila entre infinitos valores, es decir, se alcanza un comportamiento caótico. Veámoslo:

```python
# Se llama a la función para que muestre la gráfica con C=3.6.  
logistic_equation(3.6, 51, 0.85)
```

```python
# Se llama a la función para que muestre la gráfica con C=4.  
logistic_equation(4, 51, 0.85)
```

Este comportamiento se conoce como *atractores extraños*. Por lo tanto, podemos resumir que a medida que C va creciendo se observan tres comportamientos de distintos tipos de atractores: puntos fijos, oscilación periódica y atractores extraños. También hemos visto como a medida que aumentamos el valor de C, iterando la ecuación logística para valores dados de C primero se alcanzan puntos fijos, luego oscilaciones de período 2, luego oscilaciones de período 4, luego período 8, y así sucesivamente hasta que se alcanza el caos. En la teoría de sistemas dinámicos, cada uno de estos cambios de comportamiento se conoce como *bifurcación*. Estas bifurcaciones se pueden sumarizar en el conocido *diagrama de bifurcaciones*, en el que se representa el atractor al que tiende la ecuación logística para cada valor de C. Pasemos a representar este diagrama de bifurcaciones:

```python
def bifurcation_diagram(xmin=0, xmax=4, n_iter=100, step=0.01):
    '''
    Función que genera el diagrama de bifurcaciones.
    '''
    
    # Se itera en el rango indicado de valores de C.
    for c in np.arange(xmin, xmax, step):
        
        # Se define un valor inicial para x.
        seed = 0.5
        
        # Se definen un conjunto y una lista que servirán para ir metiendo los 
        # diferentes valores de x que se van obteniendo al ir iterando.
        xs_set = set() 
        xs_list = []
        
        # Se evalúan los casos en que no se alcanza el caos.
        if c < 3.569946:
            
            # Se inserta el primer valor de x, x0, en el conjunto y la lista.
            x = seed
            xs_list.append(x)
            xs_set.add(x)
            
            # Se itera varias veces para un valor de C dado.
            for v in range(1, n_iter):
                
                # Se calcula el nuevo valor de x.
                x = c * x * (1 - x) 
                
                # Si resulta que el nuevo valor de x ya se encuentra en el conjunto de valores x
                # encontrados hasta el momento, significa que ya se ha alcanzado el atractor.
                if x in xs_set:
                    
                    # Se guarda en una lista o bien el punto fijo o bien los valores distintos 
                    # entre los que oscila el atractor.
                    xs_unicos = xs_list[xs_list.index(x):v+1]
                    
                    # Se genera una lista con el valor de C repetido tantas veces como valores distintos
                    # existen entre los que oscila el atractor.
                    cs = [c for i in range(len(xs_unicos))]
                    
                    # Se representan gráficamente los valores entre los que oscila el atractor para el 
                    # valor de C considerado.
                    plt.scatter(cs, xs_unicos, color='blue', marker='s', s=1) 
                    
                    # Se sale del bucle.
                    break
                
                # Se añade el valor nuevo de x a la lista y el conjunto.
                xs_list.append(x)
                xs_set.add(x)
        
        # Se evalúan los casos en que no sí se alcanza el caos.
        else:
            
            # Se itera para encontrar hasta 1000 puntos sobre los que oscila el atractor caótico.
            for _ in range(1000):
                
                # Se calcula el nuevo valor de x.
                x = c * x * (1 - x) 
                
                # Se añade el valor nuevo de x a la lista.
                xs_list.append(x)
            
            # Se genera una lista con el valor de C repetido 1000 veces 
            cs = [c for i in range(1000)]
            
            # Se representan gráficamente los valores entre los que oscila el atractor para el 
            # valor de C considerado.
            plt.scatter(cs, xs_list, color='red', marker='s', s=1) 
    
    # Se establecen unos límites a los ejes.
    plt.xlim(xmin, xmax)
    plt.ylim(-0.1, 1.1)
    
    # Se establecen etiquetas a los ejes.
    plt.xlabel('C')
    plt.ylabel('x')
    
    # Se añade un título al gráfico.
    plt.title('Bifurcation diagram')
    
    # Se añade un grid al gráfico.
    plt.grid(True)
    
    # Se muestra el gráfico.
    plt.show()
    
# Se realiza una llamada a la función para representar el diagrama de bifurcaciones.    
bifurcation_diagram()
```

Aparece una especie de diagrama de bifurcaciones, pero con algunas zonas vacías, en las que no se representa el valor del atractor al que se tiende para algunos valores de C. Intuyo que esto debe ser debido a que la función que he construido itera hasta un número máximo de veces para alcanzar el atractor, pero puede ser que para estos valores de C que no aparecen sean necesaria más iteraciones para alcanzarlo. Probemos con un número de iteraciones más alto, por ejemplo, 1000, en vez de 100.

```python
# Se realiza una llamada a la función para representar el diagrama de bifurcaciones,
# aumentando el número de iteraciones.
bifurcation_diagram(n_iter=1000)
```

La suposición era correcta. Ahora aparece la representación más completa, pero aún no del todo, lo cual indica que para algunos valores, vemos que cercanos a 1.0 y a 3.0, se deben necesitar más iteraciones que 1000. Vemos también que el coste computacional es más alto, la función tarda más tiempo en realizar los cálculos. Probemos con 10000.


```python
# Se realiza una llamada a la función para representar el diagrama de bifurcaciones,
# aumentando aún más el número de iteraciones.
bifurcation_diagram(n_iter=100000)
```

Tarda bastante más tiempo, pero el resultado vale la pena, ya que ahora se ve la representación casi del todo completa. Pasemos a analizarla. Se observa como para valores de C entre 0 y 1 el atractor en un punto fijo, concretamente el valor 0. Para valores de C entre 1 y 3 el atractor va creciendo, tomando valores entre 0 y aproximadamente 0.65. Entre valores de C entre 3 y algo menos de 3.4 el atractor es doble. A partir de aquí se hace difícil ver lo que ocurre entre valores de C entre 3.5 y 4, por lo que representaré solamente este rango:

```python
# Se realiza una llamada a la función para representar el diagrama de bifurcaciones,
# acotando el rango de valores de C para visualizar mejor la zona caótica y la 
# inmediatamente anterior.
bifurcation_diagram(xmin=3.4, xmax=4, n_iter=10000, step=0.001)
```

Se como se van sucediendo los atractores, pasando a ser dobles, cuaduples y octuples. El nivel de detalle no permite ver más, pero casi inmediatamente se entra en la zona caótica.

Pasemos a hora a ver una cuestión que hemos detectado anteriormente. Hemos visto como al acercarnos a ciertos valores de C, crece el número de iteraciones necesarias para alcanzar el atractor. Intentemos visualizarlo representando el número de iteraciones necesarias para alcanzar el atractor en función del parámetro C:

```python
def atractores(xmin=0, xmax=4, step=0.005, n_iter=10000):
    '''
    Función que representa gráficamente el número de iteraciones necesarias para 
    alcanzar el atractor en función del parámetro C.
    '''
    
    # Se itera sobre el rango de valores C.
    for c in np.arange(xmin, xmax, step):
        
        # Se define un valor inicial para x.
        x = 0.5
        
        # Se define un conjunto para ir metiendo los valores que se van calculando para x.
        xs_set = set()
        
        # Se añade el primer valor de x en el conjunto.
        xs_set.add(x)
        
        # Se itera el número de veces indicado por el parámetro n_iter.
        for iteration in np.arange(1, n_iter):
            
            # Se calcula el nuevo valor de x.
            x = c * x * (1 - x) 
            
            # En el caso que el nuevo valor de x ya se encuentre en el conjunto sabemos que
            # ya hemos alcanzado el atractor.
            if x in xs_set:
                
                # Se representa el número de iteraciones para el valor de C correspondiente.
                plt.scatter(c, iteration, color='blue', marker='s', s=1)
                
                # Se sale del bucle iterativo.
                break
                
            # Se añade el nuevo valor de x en el conjunto.
            xs_set.add(x)
            
    # Se añaden etiquetas a los ejes.
    plt.xlabel('C')
    plt.ylabel('Iterations')
    
    # Se añade un título.
    plt.title('Iterations vs C')
    
    # Se añade un grid.
    plt.grid(True)
    
    # Se muestra el gráfico.
    plt.show()

    
# Se llama a la función para que muestre un gráfico del número de iteraciones 
# necesarias para alcanzar el atractor en función del parámetro C.
atractores()
```

Podemos ver como al acercarnos a los valores de C que generan bifurcaciones el número de iteraciones crece asintóticamente, para caer al sobrepasar estos valores. Veamos un zoom alrededor de la bifurcacion correspondiente al valor C=3:

```python
# Se llama a la función para que muestre un gráfico del número de iteraciones 
# en función del parámetro C, en un rango alrededor de la bifurcación en C=3.
atractores(xmin=2.995, xmax=3.005, step=0.00001, n_iter=100000)
```

Vemos que incluso ajustando un número máximo de iteraciones a 100000 no se alcanza el límite al cual tiende este valor, por lo que no considero demasiado aventurado suponer que el límite es el infinito.

### Pregunta teórica
En el código de ejemplo que podéis encontrar aquí (http://matplotlib.org/1.3.1/examples/mplot3d/lorenz_attractor.html), comentad una a una qué hace cada línea. **(0.5 puntos)**

```python
# Plot of the Lorenz Attractor based on Edward Lorenz's 1963 "Deterministic
# Nonperiodic Flow" publication.
# http://journals.ametsoc.org/doi/abs/10.1175/1520-0469%281963%29020%3C0130%3ADNF%3E2.0.CO%3B2
#
# Note: Because this is a simple non-linear ODE, it would be more easily
#       done using SciPy's ode solver, but this approach depends only
#       upon NumPy.

# Se importan las librerías necesarias.
# Se importa la librería numpy y se le asigna np como nickname.
import numpy as np
# Se importa la librería matplotlib.pyplot y se le asigna plt como nickname.
import matplotlib.pyplot as plt
# Se importa la clase Axes3D desde la librería mpl_toolkits.mpot3d, que permite representar gráficos en 3D.
from mpl_toolkits.mplot3d import Axes3D

# Se define la función lorenz, que admite 6 parámetros:
#      x, y, z: Valores que definen un punto en el espacio.
#      s: Parámetro conocido como Número de Prandtl. Suele tomarse s=10, y por este motivo se establece
#         este valor por defecto.
#      r: Parámetro conocido como número de Rayleigh. El sistema exhibe un comportamiento caótico para r= 8 
#         pero muestra órbitas periódicas para otros valores de r.
#      b: Parámetro que suele tomarse con valor b=8/3=2.667.
def lorenz(x, y, z, s=10, r=28, b=2.667):
    # Se calcula la derivada de x respecto al tiempo, dx/dt.
    x_dot = s*(y - x)
    # Se calcula la derivada de y respecto al tiempo, dy/dt.
    y_dot = r*x - y - x*z
    # Se calcula la derivada de z respecto al tiempo, dz/dt.
    z_dot = x*y - b*z
    # Se retorna una tupla con los tres valores calculados: dx/dt, dy/dt y dz/dt.
    return x_dot, y_dot, z_dot

# Se define un valor para el factor multiplicativo en la evolución de los puntos sucesivos.
dt = 0.01
# Se define un número de steps.
stepCnt = 10000

# Need one more for the initial values
# Se definen tres vectores que almacenarán las tres componentes de los puntos sucesivos que se calcularán.
# Se define un vector para los valores iniciales de x, con una longitud igual al número de steps dado por 
#     la variable stepCnt más uno.
xs = np.empty((stepCnt + 1,))
# Se define un vector para los valores iniciales de y, con una longitud igual al número de steps dado por 
#     la variable stepCnt más uno.
ys = np.empty((stepCnt + 1,))
# Se define un vector para los valores iniciales de z, con una longitud igual al número de steps dado por 
#     la variable stepCnt más uno.
zs = np.empty((stepCnt + 1,))

# Setting initial values
# Mediante una tupla se dan valores iniciales para el primer valor de x, y, z, es decir, el punto inicial.
xs[0], ys[0], zs[0] = (0., 1., 1.05)

# Stepping through "time".
# Se itera sobre un bucle que simula el paso del tiempo.
for i in range(stepCnt):
    # Derivatives of the X, Y, Z state
    # Se calculan las derivadas de x, y, z respecto al tiempo, mediante la función lorentz, 
    #     a partir de los valores que se encuentran en los índices i de los vectores para 
    #     las x, y, z, es decir, en los vectores xs, ys y zs.
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    
    # Se calculan las coordenadas del siguiente punto el siguiente valor en el vector xs, sumándole a cada 
    #     coordenada del punto anterior la derivada calculada por lafunción lorentz multiplicada por el factor dt.
    # Se calcula el valor de la coordenada x del siguiente punto.
    xs[i + 1] = xs[i] + (x_dot * dt)
    # Se calcula el valor de la coordenada y del siguiente punto.
    ys[i + 1] = ys[i] + (y_dot * dt)
    # Se calcula el valor de la coordenada z del siguiente punto.
    zs[i + 1] = zs[i] + (z_dot * dt)

# Se define un gráfico mediante la función plt() de la librería matplotlib.pyplot.
fig = plt.figure()
# Se obtienen los ejes del gráfico que se desea representar. El argumento projection='3d' indica el tipo de ejes que
#     se quieren usar.
ax = fig.gca(projection='3d')

# Se representan todos los puntos, a partir de sus coordenadas  x, y, z contenidas en los vectores xs, ys, zs.
ax.plot(xs, ys, zs)
# Se añade una etiqueta al eje de las x.
ax.set_xlabel("X Axis")
# Se añade una etiqueta al eje de las y.
ax.set_ylabel("Y Axis")
# Se añade una etiqueta al eje de las z.
ax.set_zlabel("Z Axis")
# Se añade un título al gráfico.
ax.set_title("Lorenz Attractor")

# Se muestra el gráfico.
plt.show()
```
