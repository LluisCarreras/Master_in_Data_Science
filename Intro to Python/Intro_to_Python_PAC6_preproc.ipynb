__b) ¿Cuántas muestras tienen al menos un valor perdido en cualquiera de sus atributos?__

Construímos una función que devuelve el número de muestras que tienen al menos un valor dado por el parámetro *n* de valores perdidos.

```python
# Respuesta

def num_cols_with_n_or_or_more_nulls(n):
    
    # Inicializamos el resultado a zero.
    result = 0
    
    # Iteramos sobre el resultado de data.isnull().values, que es una lista de listas,
    # en la que cada una de las listas internas corresponde a una muestra y contiene 
    # True o False para cada columna, dependiendo si para esa muestra y atributo el valor
    # es un valor perdido o no.
    for row in data.isnull().values:
        
        # Si la suma de valores perdidos en la muestra es mayor o igual al parámetro n, 
        # el resultado que se devolverá se incrementa en una unidad.
        if sum(row) >= n:
            result += 1
    
    # Se retorna el número de muestras que contienen como mínimo n valores perdidos.
    return result
```

Veamos el resultado que devuelve la función para el caso en que n=1, es decir, devuelve el número de muestras que tienen al menos 1 valor perdido en sus atributos.

```python
print num_cols_with_n_or_or_more_nulls(1)
```

Veamos el resultado que devuelve la función para el caso en que n=3, es decir, devuelve el número de muestras que tienen al menos 3 valor perdido en sus atributos.

```python
# Respuesta
print num_cols_with_n_or_or_more_nulls(3)
```

## Ejercicio 3

__Una de las columnas del conjunto de datos contiene información totalmente redundante. Describid cómo detectar qué columna es y realizad el proceso de detección. Eliminad la columna del conjunto de datos.__

__Pista: quizás la función [_corr_](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.corr.html) de pandas puede ser de utilidad.__

Utilizaré la función *corr()* de pandas, que calcula la correlación entre todos los pares de columnas. Veamos los resultados:

```python
# Respuesta
correlations = data.corr()
correlations
```

Una inspección ocular de los resultados nos permite detectar que existe una correlación de 1.0 entre las columnas *__total sulfur dioxide__* y *__total dioxide__*, lo que nos permite concluir que las dos nos dan la misma información, por lo que podemos eliminar una de las dos.

Ahora bien, sería conveniente automatizar este resultado, es decir, intentar conseguir que el sistema nos informe del hecho de que hay atributos que están totalmente correlacionados:

```python
# Se guardan en una variable tipo lista los nombres de las columnas.
atributes_names = list(data)

# Se guarda en una variable de tipo entero el número de columnas.
atributes_len = len(atributes_names)

# Se define una variable de tipo lista en la que se irán guardando los eventuales
# pares de atributos que estén completamente correlacionados.
correlated_pairs = []

# Se itera sobre todas los índices de las filas.
for row in range(atributes_len):
    
    # Se itera sobre los índices de las columnas, de manera que se eviten repeticiones de
    # pares de atributos y también se avite evaluar correlaciones de atributos consigo mismos.
    for col in range(row - 1):
        
        # Se guarda en una variable el valor que se recupera para la correlación entre los atributos
        # localizados en los índices dados por row y col.
        corr = correlations.iat[row, col]
        
        # Se evalúa si la correlación eś igual a 1.0 o a -1.0.
        if (corr == 1.0) | (corr == -1):
            
            # Si la correlación es igual a 1.0 o -1.0, se añade una lista en la lista de pares de atributos
            # correlacionados, que contiene la correlación (1.0 o -1.0), y el nombre de los dos atributos
            # correlacionados.
            correlated_pairs.append([corr, atributes_names[row], atributes_names[col]])
            
# Imprimimos la lista con los pares de atributos correlacionados
print correlated_pairs
            
```

Vemos que hemos obtenido el resultado esperado, una lista con el par de atributos *__total dioxide__* y *__total sulfur dioxide__*, con una correlación de 1.0.

Pasamos ahora a eliminar una de las dos columnas correlacionadas. Elijo eliminar la columna *__total dioxide__*.

```python
print data.shape
data = data.drop(['total dioxide'], axis=1)
print data.shape
```

__Cread 4 nuevos atributos binarios (`sulphates_Q1`, `sulphates_Q2`, `sulphates_Q3`, `sulphates_Q4`) que indiquen en qué cuartil se encuentra el valor de sulfatos (`sulphates`) de cada muestra. Así, para una muestra con un valor de sulfatos en el segundo cuartil, el atributo `sulphates_Q2` será 1 y los atributos `sulphates_Q1`, `sulphates_Q3` y `sulphates_Q4` serán 0.__

Primero, veamos un sumario estadístico de *sulphates*:

Pasemos ahora a guardar los valores que definen los 4 cuartiles en una lista:

```python
# Respuesta

# Creamos una lista vacía para que contenga los valores que definen los cuartiles.
sulphate_quartiles = []

# Iteramos sobre los porcentajes que definen los cuartiles.
for quantile in [0.00, 0.25, 0.50, 0.75, 1.00]:
    # Para cada valor de porcentaje, calculamos su cuartil.
    sulphates_Q_value = data['sulphates'].quantile(quantile)
    # Imprimimos el valor del cuartil obtenido.
    print "Q%d: %.2f" % (int(4 * quantile), sulphates_Q_value)
    # Añadimos el cuartil obtenido a la lista de cuartiles
    sulphate_quartiles.append(sulphates_Q_value)
    
# Imprimimos la lista de cuartiles.
print sulphate_quartiles
```

Una vez tenemos los valores que definin los cuartiles, pasemos a añadir al dataframe las 4 nuevas columnas y las rellenamos para tener los valores que se indican en el enunciado. Podemos realizar esto a partir de una función que creamos.

```python
# Importamos la librería operator para poder usar la operación "and_" y así evitar problemas que surgen al aplicar 
# operadores booleanos a Series o Dataframes.
import operator

# Creamos una función que nos añade las columnas requeridas por el enunciado.
def create_quantile_column(df, quartiles_list):
    # Iteramos sobre la lista de cuartiles. Utilizamos la función enumerate() para poder trabajar con los valores
    # y con los índices de estos valores.
    for idx, quartil in enumerate(quartiles_list):
        # Para calcular cada columna usmos el cuartil del índice en el que estamos iterando y el siguiente. No iteramos sobre el 
        # último índice ya que este no tiene índice siguiente.
        if idx < len(quartiles_list)-1:
            # Construimos un string con el nombre de la columna en cuestión.
            col_name = 'sulphates_Q%d' % (idx+1)
            # Construimos la columna. Para ello, creamos una columna a la que nombramos seg´un el nombre que tenemos en la variable
            # "col_name", y aplicamos la función "and_" de la librería "operator", que nos devolverá el valor de la columna "sulphates"
            #  en las filas que cumplen que el valor de la columna "sulphates" esté comprendido entre los cuartiles correspondientes. 
            # En los otros casos pondrá un valor NaN. Al aplicarle la función "astype(bool) lo que hacemos es convertir los valores
            # numéricos en booleanos, y como todos son diferentes de 0, pasarán a ser 1.
            data[col_name] = data.loc[operator.and_(data['sulphates'] >= sulphate_quartiles[idx], data['sulphates'] < sulphate_quartiles[idx+1]), 
                                            ['sulphates']].astype(bool)
            # Los valores NaN pasan a ser False.
            data[col_name] = data[col_name].fillna(False)
            # Ahora toda la columna son o bien False o True. Pasamos estos valores a sus correspondientes valores numéricos, 0 o 1.
            data[col_name] = data[col_name].astype(int)
 
# Aplicamos la función construida a nuestro dataframe y lista de cuartiles.            
create_quantile_column(data, sulphate_quartiles)
# Mostramos las primeras 20 filas.
data.head(20)
```
